```python
import pandas as pd
import glob

# Get a list of all files with a specific pattern (e.g., all_data*.txt)
file_pattern = "all_data.txt"
files = glob.glob(f"{my_files}/{file_pattern}")

# Initialize an empty list to store individual DataFrames
dfs = []

# Initialize an empty set to store unique groups
all_groups = set()

# Loop through each file and read its data into a DataFrame
for file in files:
    raw_df = pd.read_csv(file, sep='\t', index_col=0)

    # Update the set with unique groups in each file
    all_groups.update(raw_df['group'].unique())

    # Append the DataFrame to the list
    dfs.append(raw_df)

# Print all unique groups
print("All unique groups:", all_groups)

# Concatenate all DataFrames in the list into a single DataFrame
combined_df = pd.concat(dfs, ignore_index=True)

# Print the length and the first 10 rows of the combined DataFrame
print(len(combined_df))
print(combined_df.head(10))
```
